```python
import pandas as pd

df = pd.read_csv("filled_data.csv")

df
```

```python
df_grouped = df.groupby("Customer_ID")
```

```python
cols_to_drop = ['Unnamed: 0.3', 'Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0', 'ID']
df_viz_1 = df.drop(cols_to_drop, axis=1)
df_viz_1
```

```python
# Assuming your DataFrame is named df
import re

# Step 1: Extract unique types of loans from the "Type_of_Loan" column
unique_loan_types = set()

for col_str in [i for i in df_viz_1["Type_of_Loan"].unique()]:
    types = re.split(', |and ', str(col_str))
    #types = str(col_str).split(', | and ')
    unique_loan_types.update(set(types))

unique_loan_types.remove('')
unique_loan_types.remove('nan')
# Step 2: Convert the set to a list
loan_types = list(unique_loan_types)

# Now, loan_types contains all unique types of loans
print(set(loan_types))
```

```python
# Step 1: Create new columns for each type of loan
for loan_type in loan_types:
    df_viz_1[loan_type] = 0

# Step 2: Update the new columns based on the "Type_of_Loan" column
for index, row in df_viz_1.iterrows():
    type_of_loans = re.split(', |and ', str(row['Type_of_Loan']))
    for loan_type in loan_types:
        if loan_type in type_of_loans:
            df_viz_1.at[index, loan_type] = 1

# Step 3: Drop the original "Type_of_Loan" column
df_viz_1 = df_viz_1.drop('Type_of_Loan', axis=1)

df_viz_1.rename(columns = {'Not Specified':'Loan Not Specified'}, inplace = True)

# Now, df contains one-hot encoded columns for each type of loan
df_viz_1.head(10)
```

# VIZ 1

```python

# Assuming your DataFrame is named df

# Group by Customer_ID and aggregate
agg_df = df_viz_1.groupby('Customer_ID').agg({
    'Monthly_Inhand_Salary': 'mean',
    'Total_EMI_per_month': 'mean',
    'Loan Not Specified': 'max',  # If any entry is 1, the max will be 1
    'Mortgage Loan': 'max',
    'Home Equity Loan': 'max',
    'Auto Loan': 'max',
    'Student Loan': 'max',
    'Personal Loan': 'max',
    'Payday Loan': 'max',
    'Debt Consolidation Loan': 'max',
    'Credit-Builder Loan': 'max'
}).reset_index()

# Convert values to 0 or 1
loan_columns = ['Loan Not Specified', 'Mortgage Loan', 'Home Equity Loan', 'Auto Loan',
                'Student Loan', 'Personal Loan', 'Payday Loan', 'Debt Consolidation Loan',
                'Credit-Builder Loan']

agg_df[loan_columns] = (agg_df[loan_columns] > 0).astype(int)

# Now, agg_df contains the desired information
agg_df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your aggregated DataFrame is named agg_df

# Define the colors for each loan type
colors = sns.color_palette("husl", n_colors=len(loan_columns))

# Create a scatter plot for each loan type
for i, loan_type in enumerate(loan_columns):
    # Select customers with the specific loan type
    subset_df = agg_df[agg_df[loan_type] == 1]

    # Scatter plot
    plt.scatter(subset_df['Monthly_Inhand_Salary'], subset_df['Total_EMI_per_month'], label=loan_type, color=colors[i])

# Set labels and title
plt.xlabel('Mean Monthly Inhand Salary')
plt.ylabel('Mean Total EMI per Month')
plt.title('Scatter Plot of Monthly Inhand Salary vs Total EMI per Month by Loan Type')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your aggregated DataFrame is named agg_df


# Create a new DataFrame for plotting
plot_df = pd.DataFrame(columns=['Loan Type', 'Mean Monthly Inhand Salary', 'Mean Total EMI per Month'])

# Define the colors for each loan type
colors = sns.color_palette("husl", n_colors=len(loan_columns))

# Populate the plot DataFrame
for i, loan_type in enumerate(loan_columns):
    # Select customers with the specific loan type
    subset_df = agg_df[agg_df[loan_type] == 1]

    # Calculate means
    mean_salary = subset_df['Monthly_Inhand_Salary'].mean()
    mean_emi = subset_df['Total_EMI_per_month'].mean()

    # Append to the plot DataFrame
    plot_df = plot_df.append({'Loan Type': loan_type,
                              'Mean Monthly Inhand Salary': mean_salary,
                              'Mean Total EMI per Month': mean_emi}, ignore_index=True)

# Sort the plot_df DataFrame alphabetically by 'Loan Type'
plot_df = plot_df.sort_values(by='Loan Type')
    
# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mean Monthly Inhand Salary', y='Mean Total EMI per Month', hue='Loan Type', data=plot_df, s=100, palette=colors)

# Set labels and title
plt.xlabel('Mean Monthly Inhand Salary')
plt.ylabel('Mean Total EMI per Month')
plt.title('Scatter Plot of Monthly Inhand Salary vs Total EMI per Month by Loan Type')

# Add legend
plt.legend()

plt.savefig("salary-EMI.jpg", bbox_inches='tight', dpi=300)

# Show the plot
plt.show()

```
